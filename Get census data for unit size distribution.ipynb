##2 Using the native API (fewer dependencies)
This other method has only minimal dependencies

```python
import pandas as pd
import requests, json
import urllib2
apikey = os.environ["CENSUS"] ## replace if needed with a string
```

```python
## get bay area fips codes
#states.CA.fips
#fips = pd.read_csv('/home/aksel/Documents/Data/ca_county_fips.csv',sep='\t')
#fips['fipscode']=fips.ix[:,'FIPS Code'].apply(lambda x: str(x).zfill(3))
#fips.index=fips.fipscode
#fips.ix[fips.bayarea==1,'fipscode'].tolist()
#bayareafips = fips.ix[fips.bayarea==1,'fipscode'].tolist()
bayareafips ={'001':'ala', '013':'con', '041':'mar', '055':'nap', '075':'sfr', '081':'san', '085':'scl', '095':'son', '097':'sol'}
bayareafips.keys()
```

```python
#%debug
def get_data(year,dataset,table):
    """
    Gets data from api based on year, dataset and tablename
    """
    
    import re
    #concatenate table list to pass to url param
    stable = ','.join(table)
    
    url = 'http://api.census.gov/data/%(yr)s/%(dset)s?key=%(key)s&get=%(tbl)s&for=tract:*&in=state:06'
    url = url %{
    'key':apikey,
    'yr':year,
    'dset':dataset,
    'tbl':stable
    }
    print url
    ## get data
    resp = requests.get(url)
    
    ## get variable details, labels
    jsontext = requests.get('http://api.census.gov/data/2012/acs5/variables.json')
    
    try:
        data = pd.read_json(resp.text)
        
        ## col names are found at index 0
        cols = data.ix[0,].tolist()
           
        data.columns=cols
                        
        ## data ships with empty row--remove
        data=data.drop(data.index[:1])
        
        ## geo stuff
        ##    keep bay area counties only
        data = data.ix[data.county.isin(bayareafips.keys()),:]
        
        ## add unique geoid as index 
        data.index = data.state+data.county+data.tract
        
        ## add county abbreviation
        data['county'] = data.county.map(bayareafips)
        data.drop(['state','tract'],inplace=True,axis=1)
        
        ## make sure numeric data is numeric
        data.ix[:,table]=data[table].astype('float64')
        
        ## rename columns, looking up in dictionary. Perhaps bad form w verbose names
        datadict = pd.DataFrame(jsontext.json()['variables']).T
        
        ## get labels, stuff in dict
        datadict_rename = datadict.ix[table,'label'].to_dict()
        
        ## rename
        data.rename(columns=datadict_rename,inplace=True)
        
        fname = "census_%s_%s_%s.csv" %(dataset,year,re.sub('\.  | ','_',''.join(datadict.ix[datadict.index==table[0],:].concept.tolist())))
        print "writing data to %s" %fname
        
        data.to_csv(fname)
        return data
    except Exception, e:
        raise    # decoding failed
        print "holy smokes!"
        print "failed on %s" %url
    else:
        pass
```

```python
import re, io, numpy as np
from cStringIO import StringIO 
    
def map_value_labels(codes_pasted,var):
    """
    copy id:value directly from data dict
    """
    
    ## subst out dots and leading space
    codes_pasted = re.sub(r' \.','\t', codes_pasted)
    
    ## read code, value to dataframe
    df_codes = pd.read_csv(StringIO(codes_pasted), 
                           encoding='utf8', sep="\t",
                           header=0,names=['id','value'],
                           dtype={'id': object, 'value': object})
    
    ## set index=id so we can make json file for mapping to pums values
    df_codes.index=df_codes.id.astype(np.int64)
    df_codes=df_codes.ix[:,['value']]
    
    ## populate a dictionary {id:value} we can use for mapping values
    code_mapping_dict=dict()
    for row in df_codes.itertuples(): code_mapping_dict[row[0]]=row[1]
    
    return code_mapping_dict

def assign_value_labels(df,value_map):
    
    ##prepare to write encoding back to data
    target = "%s_l"%var
    df[target] = df[var].map(value_map)
    #pums_p['RAC1P'].astype(object).map(df_codes.to_json)
    print df[target].head()
```
